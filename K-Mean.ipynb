```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Final_data.csv')

# Selecting relevant features for clustering
# Assuming 'features' variable contains the selected features from your dataset
#features = data[['Feature1', 'Feature2', 'Feature3', 'Feature4']]  # Replace with your actual feature columns

features = data[['Monthly SKUs Purchased', 'Monthly Vendors Purchased', 'Monthly Number of Sales', 'Monthly Sum Sales ($)', 'Monthly Gross Margin ($)']]
target = data['Churned']


# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Reducing the features to 2D for visualization (if more than 2 features)
pca = PCA(n_components=2)
features_2D = pca.fit_transform(features_scaled)
```

```python
# Number of clusters
k = 3  # Adjust this based on your requirements

# Applying KMeans
kmeans = KMeans(n_clusters=k, random_state=0)
clusters = kmeans.fit_predict(features_2D)
```

```python
# Scatter plot of the clusters
plt.figure(figsize=(8, 6))
plt.scatter(features_2D[:, 0], features_2D[:, 1], c=clusters, cmap='viridis', marker='o')

# Marking the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')

plt.title('K-Means Clustering')
plt.xlabel('Feature 1 (PCA)')
plt.ylabel('Feature 2 (PCA)')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# Load the dataset
data = pd.read_csv('Final_data.csv')

# Selecting relevant numerical features for clustering
features = data[['Monthly SKUs Purchased', 'Monthly Vendors Purchased', 
                 'Monthly Number of Sales', 'Monthly Sum Sales ($)', 'Monthly Gross Margin ($)']]

# Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Reducing the features to 2D for visualization
pca = PCA(n_components=2)
features_2D = pca.fit_transform(features_scaled)

# Number of clusters
k = 3  # Adjust based on your requirements

# Applying KMeans
kmeans = KMeans(n_clusters=k, random_state=0)
clusters = kmeans.fit_predict(features_2D)

# Visualization
plt.figure(figsize=(10, 8))
plt.scatter(features_2D[:, 0], features_2D[:, 1], c=clusters, cmap='viridis', marker='o')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title('K-Means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()

# Classification Report (if applicable)
# Replace 'true_labels' with the actual column name of true labels in your dataset
# true_labels = data['YourTrueLabelColumn']
# print(classification_report(true_labels, clusters))
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(features_scaled, clusters)
print(f"Silhouette Score: {silhouette_avg:.2f}")
```

```python
sse = [] 
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(features_2D)
    sse.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()
```
